## Scrape Organizations in United States

```
browser.get("https://www.coursera.org/about/partners/us")

count = len(browser.find_elements_by_class_name('rc-PartnerBox'))
```

```
univerisities = defaultdict(list)

for i in range(0, count):
    
    organizations = browser.find_elements_by_class_name('rc-PartnerBox')
    
    name = organizations[i].text

    url = organizations[i].find_element_by_tag_name('a')
    url = url.get_attribute('href')

    if 'University' in str(name):
      
      univerisities['Name'].append(name)
      univerisities['URL'].append(url)
    
    elif 'Institute' in str(name):

      univerisities['Name'].append(name)
      univerisities['URL'].append(url)

    elif 'College' in str(name):

      univerisities['Name'].append(name)
      univerisities['URL'].append(url)

    elif 'School' in str(name):

      univerisities['Name'].append(name)
      univerisities['URL'].append(url)

    else:
      continue
```

```
univerisities = pd.DataFrame(univerisities)
univerisities.to_csv('universities.csv')
univerisities
```

## Scrape Instructors

```
instructors = defaultdict(list)

for row in univerisities.itertuples(index=False):
  
  if 'coursera' not in row[1]:
    continue
  
  else:
  
    browser.get(row[1])
    instructor_count = len(browser.find_elements_by_class_name("rc-Instructor"))

    for i in range(0, instructor_count):

      instructor_info = browser.find_elements_by_class_name("rc-Instructor")
      
      image = instructor_info[i].find_element_by_class_name('instructorPhoto')

      instructors['Image'].append(image.find_element_by_tag_name('a').find_element_by_tag_name('img').get_attribute('src'))
      
      info = instructor_info[i].find_element_by_class_name('instructorDetails')
      
      instructors['Name'].append(info.find_element_by_tag_name('h4').text)
      instructors['URL'].append(info.find_element_by_tag_name('h4').find_element_by_tag_name('a').get_attribute('href'))
      instructors['Headline'].append(info.find_element_by_class_name('body-1-text').text)
      instructors['Affiliation'].append(row[0])
```

```
instructors = pd.DataFrame(instructors)
instructors.to_csv("instructors.csv")
instructors
```

## Scrape Instructors and Courses

```
courses = set()
instructor_courses = defaultdict(list)

instructor_bios = []
instructor_external_links = []

for row in instructors.itertuples(index = False):
  
  browser.get(row[2])
  
  #Bio
  instructor_bios.append(browser.find_element_by_class_name("Col_i9j08c-o_O-smCol12_s10s1s-o_O-mdCol12_vi0cxf").text)

  #External Links
  try:
    
    external_links = browser.find_element_by_class_name("externalLinks_1uyl3cv")
    links = external_links.find_elements_by_tag_name('li')

    temp = []

    for link in links:
      temp.append(link.find_element_by_tag_name('a').get_attribute('href'))
    
    instructor_external_links.append(temp)
  
  except:
    instructor_external_links.append(None)

  #Courses
  instructor_course_info = browser.find_element_by_class_name("Row_nvwp6p")
  course_info = instructor_course_info.find_elements_by_class_name('Col_i9j08c-o_O-smCol6_1g8zirr-o_O-mdCol4_cuxg7k-o_O-lgCol3_1a7wv9')

  for course in course_info:

    link = course.find_element_by_tag_name('a').get_attribute('href')

    courses.add(link)
    instructor_courses['Instructor'].append(row[2])
    instructor_courses['Course'].append(link)
```

```
instructor_courses = pd.DataFrame(instructor_courses)
instructor_courses.to_csv("instructor_courses.csv")
instructor_courses
```

```
instructors['External_Links'] = instructor_external_links
instructors['Bio'] = instructor_bios
instructors.to_csv("instructors.csv")
instructors
```

## Scrape Course Information

```
course_data = pd.DataFrame()

for idx, link in enumerate(courses):

  temp = dict()

  temp['Course ID'] = str(idx)
  temp['Course URL'] = link

  browser.get(link + '/reviews')
  #print(idx, link + '/reviews')

  course_name = ' '

  try:
    temp['Course Name'] = course_name.join(browser.find_element_by_class_name('backToCDPLink_15q6gcm').text.split(' ')[2:])

  except:
    temp['Course Name'] = None

  #About Course

  body = browser.find_element_by_class_name('AboutCourse')

  ##Check 'View All' button

  try:

    button = body.find_element_by_tag_name('button')
    button.click()
    time.sleep(0.1)

  except:

    button = None

  temp['About the Course'] = body.find_element_by_class_name('m-b-2').text

  #Ratings
  try:
  
    temp['Rating'] = float(browser.find_element_by_class_name("H4_1k76nzj-o_O-weightBold_uvlhiv-o_O-bold_1byw3y2").text.split('\n')[0])
    temp['Number of Ratings'] = int(browser.find_element_by_class_name("P_gjs17i-o_O-weightNormal_s9jwp5-o_O-fontBody_56f0wi").text.split(' ')[0].replace(',',''))

  except Exception as error:
    
    temp['Rating'] = None
    temp['Number of Ratings'] = None
    #print(error)

  #Reviews
  try:

    temp['Number of Reviews'] = int(browser.find_element_by_class_name("reviewsCount").text.split(' ')[0].replace(',',''))
  
  except Exception as error:

    temp['Number of Reviews'] = None
    #print(error)

  course_data = course_data.append(pd.DataFrame(temp, index=[0]), ignore_index=True)
  course_data.to_csv('/content/drive/My Drive/Research Work/course_data.csv')
```

```
course_data = pd.read_csv("/content/drive/My Drive/Research Work/Coursera/course_data.csv")
```

```
for idx, row in islice(course_data.iterrows(), 256, None):

  link = row['Course URL']
  browser.get(link + '/reviews')
  print(idx)

  try:
    
    reviews = browser.find_element_by_class_name('rc-ReviewsContainer')
  
    navigation = reviews.find_element_by_tag_name('nav')
    pages = navigation.find_elements_by_tag_name('li')
  
    page_count = int(pages[-2].text)

  except:

    page_count = 0

  review_data = defaultdict(list)

  for page_number in range(1, page_count+1):

    flag = True
    
    while flag:
      
      browser.get(link + '/reviews?page=' + str(page_number))

      review_list = browser.find_element_by_class_name("rc-ReviewsList")

      try:
        
        reviews = review_list.find_elements_by_class_name("Row_nvwp6p")
        flag = False

      except Exception as error:

        print(error, link + '/reviews?page=' + str(page_number))
        reviews = []
    
    for review in reviews:

      review_info = review.find_element_by_class_name("Col_i9j08c-o_O-mdCol8_1xhmn9")
      
      review_data['Body'].append(review_info.find_element_by_class_name("reviewText").text)
      review_data['By'].append(review_info.find_element_by_class_name("reviewerName").text[3:])
      review_data['Date of Review'].append(review_info.find_element_by_class_name("dateOfReview").text)

      stars = review_info.find_elements_by_class_name("star_16s6vgr")
      star_count = 0

      for star in stars:

        if star.text == 'Filled Star':
          star_count += 1

      review_data['Stars'].append(star_count)
  
  review_data = pd.DataFrame(review_data)
  review_data.to_csv("/content/drive/My Drive/Research Work/Coursera/Reviews/" + str(idx) + ".csv")
```
